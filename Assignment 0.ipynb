```python
import numpy
from matplotlib import pyplot as plt
import pandas as pd
from mpl_toolkits.axes_grid1.axes_divider import make_axes_area_auto_adjustable
```

```python
#list of chess games with 16 features to classify the game as {white_win,black_win}
games = pd.read_csv('games.csv')
plt.plot(games.white_rating,games.black_rating,'x')
plt.title('Ratings')
plt.xlabel('White')
plt.ylabel('Black')
plt.show
```

```python
#calculates and displays average turns in wins for each color
whitecount=0
whiteturn=0
blackcount=0
blackturn=0
index = 0
for winner in games.winner:
    if winner=='white':
        whitecount+=1
        whiteturn+=games.turns.iloc[index]
    elif winner == 'black' :
        blackcount+=1
        blackturn+=games.turns.iloc[index]
    index+=1
plt.bar(['white','black'],[(whiteturn/whitecount),(blackturn/blackcount)])
plt.title('Average turns in wins')
plt.ylabel('Turns')
plt.show
```

```python
ootCount = 0
resignCount =0
mateCount=0
drawCount=0
for status in games.victory_status:
    if status == 'outoftime':
        ootCount+=1
    elif status == 'resign':
        resignCount+=1
    elif status == 'mate':
        mateCount+=1
    elif status == 'draw':
        drawCount+=1
plt.bar(['outoftime','resign','mate','draw'],[ootCount,resignCount,mateCount,drawCount])
plt.title('Types of wins')
plt.show
```

```python
#list of wines with 14 features in order to use regression to predict rating
wine = pd.read_csv('winemag-data-130k-v2.csv')
plt.plot(wine.price,wine.points,'x')
plt.title('Point distribution based on price')
plt.xlabel('Price')
plt.ylabel("Points")
plt.show
```

```python
countries = []
ratings = []
index=0
for country in wine.country:
    if str(country) not in countries:
        countries.append(str(country))
        ratings.append(wine.points.iloc[index])
    else:
        ratings[countries.index(str(country))] = (wine.points.iloc[index]+ratings[countries.index(str(country))])/2
    index+=1

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1,.8, 2])
ax.barh(countries,ratings)
ax.set_yticks(countries)
ax.set_yticklabels(countries)
plt.show
```

```python
names=[]
ratings=[]
index=0
for name in wine.taster_name:
    if str(name) not in names:
        names.append(str(name))
        ratings.append(wine.points.iloc[index])
    else:
        ratings[names.index(str(name))] = (wine.points.iloc[index]+ratings[names.index(str(name))])/2
    index+=1
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 1])
ax.barh(names,ratings)
ax.set_yticks(names)
ax.set_yticklabels(names)
plt.show
```
