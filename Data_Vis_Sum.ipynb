# Dataset Creation and Labeling

Our dataset comes from NOAA's Milky Way panoramas (https://sos.noaa.gov/datasets/milky-way-panorama-constellation-outlines/ ). Specifically the original with no extra data and the panorama with the constellation outlines overlayed. We created our dataset using a script which takes screenshots of a specific region of screen every ~.33s. We then ran the code while the panorama was auto rotating to get a 360 degree view of images. This was done 3 times at 3 different rotations about the y axis (0,45 and 90 degrees). This allowed us to get multiple angles of every constellation on the sphere. Our data is separated into two folders of images which correspond to each other. This meant we could look at the image of the panorama with the constellation overlay and create our target vectors from that. We began labeling our data by creating an excel spreadsheet (still in progress at this time) where each row corresponds to a specific image and was a set of 1s and 0s for all 81 constellations on the sphere. A value of 1 corresponded to the constellation in that column being visible in that particular image, and a value of 0 meant it was not. Once the labeling process is completed, we can train our model to classify the images with no overlays and hopefully even get our algorithm to overlay the constellations itself.
